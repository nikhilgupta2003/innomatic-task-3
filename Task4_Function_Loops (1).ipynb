## Problem 1: Mobile Recharge Validation System

```python
def validate_recharge():
    valid_plans = [199, 299, 399, 599]

    while True:
        try:
            amount = int(input("Enter recharge amount (₹): "))

            # Check minimum recharge rule
            if amount < 50:
                print("Recharge must be at least ₹50.")
                continue

            # Check if plan is valid
            if amount in valid_plans:
                print("Recharge Successful! Plan Activated.")
                break
            else:
                print("Invalid plan amount.")
        
        except ValueError:
            print("Please enter a valid numeric amount.")

        # Retry option
        retry = input("Do you want to try again? (yes/no): ").lower()
        if retry != "yes":
            print("Recharge Cancelled.")
            break


# Function call
validate_recharge()
```

## Problem 2: Inventory Reorder Alert System

```python
def check_inventory(stock_dict):
    for product, quantity in stock_dict.items():
        if quantity < 15:
            print(f"{product}: Reorder Alert")
        else:
            print(f"{product}: Stock OK")

# Example
inventory = {
    "Apples": 10,
    "Bananas": 25,
    "Oranges": 12
}
check_inventory(inventory)
```

## Problem 3: Student Result Processing System

```python
def process_result(marks):
    total = 0
    for m in marks:
        total += m
    avg = total / len(marks)
    print("Average Marks:", avg)
    if avg >= 50:
        return "Pass"
    else:
        return "Fail"

# Example
result = process_result([60, 45, 70, 55])
print("Result:", result)
```

## Problem 4: Cab Fare Estimator with Retry Option

```python
def calculate_fare():
    base_fare = 50
    per_km = 12

    while True:
        try:
            km = float(input("Enter distance (km): "))
            peak = input("Peak hour? (yes/no): ").lower()

            # Fare calculation
            fare = base_fare + (per_km * km)

            # Peak hour charges
            if peak == "yes":
                fare *= 1.25

            print(f"Estimated Fare: ₹{fare:.2f}")

        except ValueError:
            print("Please enter valid numeric distance.")

        # Retry option
        retry = input("Do you want to calculate again? (yes/no): ").lower()
        if retry != "yes":
            print("Thank you for using Cab Fare Estimator.")
            break


# Function call
calculate_fare()
```

## Problem 5: Employee Attendance Eligibility Checker

```python
def check_attendance(attendance_list):
    present = 0
    for day in attendance_list:
        if day == "P":
            present += 1

    percentage = (present / len(attendance_list)) * 100
    print("Attendance %:", percentage)

    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"

# Example
status = check_attendance(["P", "A", "P", "P", "A", "P", "P", "P"])
print("Status:", status)
```

## Problem 6: Password Strength Checker

```python
def check_password(password):
    has_digit = False
    has_special = False

    for ch in password:
        if ch.isdigit():
            has_digit = True
        if ch in "@#$":
            has_special = True

    if len(password) >= 8 and has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password"

# Example
print(check_password("Test@123"))
```
