## Problem 1: Employee Performance Bonus Eligibility

```python
# Employee Performance Bonus Eligibility
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# Find highest score
max_score = max(employees.values())

# Find employees with highest score
top_performers = [name for name, score in employees.items() if score == max_score]

print(f"Top Performers Eligible for Bonus: {', '.join(top_performers)} (Score: {max_score})")
```

## Problem 2: Search Query Keyword Analysis

```python
import string

query = "Buy mobile phone buy phone online"

# Convert to lowercase
query = query.lower()

# Remove punctuation
query = query.translate(str.maketrans('', '', string.punctuation))

# Count frequency
words = query.split()
freq = {}

for word in words:
    freq[word] = freq.get(word, 0) + 1

# Display words searched more than once
result = {k: v for k, v in freq.items() if v > 1}
print(result)
```

## Problem 3: Sensor Data Validation

```python
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

# Store valid even readings with index
valid_readings = [(i, val) for i, val in enumerate(sensor_readings) if val % 2 == 0]

print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

## Problem 4: Email Domain Usage Analysis

```python
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

domain_count = {}

for email in emails:
    domain = email.split('@')[1]
    domain_count[domain] = domain_count.get(domain, 0) + 1

total = len(emails)

# Calculate percentage
for domain, count in domain_count.items():
    percentage = (count / total) * 100
    print(f"{domain}: {percentage:.0f}%")
```

## Problem 5: Sales Spike Detection

```python
sales = [1200, 1500, 900, 2200, 1400, 3000]

# Calculate average
avg_sales = sum(sales) / len(sales)

# Detect spikes (>30% above average)
threshold = avg_sales * 1.3

for i, value in enumerate(sales, start=1):
    if value > threshold:
        print(f"Day {i}: {value}")
```

## Problem 6: Duplicate User ID Detection

```python
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

count = {}

for uid in user_ids:
    count[uid] = count.get(uid, 0) + 1

# Display duplicates only
for uid, c in count.items():
    if c > 1:
        print(f"{uid} → {c} times")
```
